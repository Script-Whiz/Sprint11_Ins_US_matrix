### KNN ( Nearest Neighbor )

```python
def get_knn(df, n, k, metric):
    
    nn = NearestNeighbors(metric=metric).fit(df)
    
    # Prepare the single data point as a DataFrame to maintain feature names
    target_row = pd.DataFrame([df.iloc[n]], columns=df.columns)
    
    nn_distances, nn_indices = nn.kneighbors(target_row, k, return_distance=True)
    
    df_res = pd.concat([
        df.iloc[nn_indices[0]], 
        pd.DataFrame(nn_distances.T, index=nn_indices[0], columns=['distance'])
        ], axis=1)
    
    return df_res
```

- param df: pandas DataFrame used to find similar objects within
- param n: object no for which the nearest neighbours are looked for
- param k: the number of the nearest neighbours to return
- param metric: name of distance metric

- KNN: Scaled dataset - Euclidean Distance

```python
#Call the KNN. The target is the 11th row.
print('Scaled dataset - Euclidean Distance')
display(get_knn(features_scaled_df, 10, 12, 'euclidean'))
```

- KNN: Scaled Dataset - Manhattan Distance

```python
#Call the KNN. The target is the 11th row.
print('Scaled dataset - Manhattan Distance')
display(get_knn(features_scaled_df, 10, 12, 'manhattan'))
```

Most of the same neighbors appear in both the Eucidean and the Manhattan output.

- KNN: NOT scaled Dataset - Euclidean

```python
ins_not_scaled = ins.copy()
```

```python
# Visual checking for NaNs
print(ins_not_scaled.isnull().sum().sort_values().tail())
```

```python
# Call the KNN function. The target is the 11th row.
knn_result_notScaled_euc = get_knn(ins_not_scaled, 10, 12, 'euclidean')
print(knn_result_notScaled_euc)
```

- KNN: NOT scaled - Manhattan

```python
# Call the KNN function. The target is the 11th row.
knn_result_notScaled_manh = get_knn(ins_not_scaled, 10, 12, 'manhattan')
print(knn_result_notScaled_manh)
```

A lot of the same neighbors show up on both of the NOT scaled KNN output, for both Euclidean and Manhattan distances.
